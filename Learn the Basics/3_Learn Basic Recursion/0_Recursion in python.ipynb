```python
def rec(cnt):

    if cnt == 4:
        return
    
    print("stmt 1",cnt)
    rec(cnt +1)
    print("stmt 2",cnt)
    
rec(0)
```

```python
def sumofn(sum,number):

    if number == 0:
        return sum
    
    sum = sum + (number % 10)
    number //= 10
    return sumofn(sum,number)

n = int(input())

sumofn(0,n)
```
