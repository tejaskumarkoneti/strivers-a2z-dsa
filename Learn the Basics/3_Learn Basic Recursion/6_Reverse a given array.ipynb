```python
def reverse(num,l,r):
    
    if l >= r:
        return
    
    num[l],num[r] = num[r],num[l]

    reverse(num,l+1,r-1)

num = [1,2,3,4,5]

r = reverse(num,0,len(num)-1)
print(r)
 
```

```python
def reverse(num,i):

    l = len(num)

    if i >= len(num)/2:
        return
    
    num[i],num[l-i-1] = num[l-i-1] , num[i]

    reverse(num,i+1)

    return num

num = [1,2,3,4,5]

r = reverse(num,0)
print(r)
 
```
