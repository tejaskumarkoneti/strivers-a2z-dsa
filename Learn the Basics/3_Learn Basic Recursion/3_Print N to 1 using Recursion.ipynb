### Input: N = 4
### Output: 4, 3, 2, 1
### Explanation: All the numbers from 4 to 1 are printed.
### Input: N = 1
### Output: 1 
### Explanation: This is the base case.

```python
def nameN(i,n):
    if i < 1:
        return
    print(i,end=" ")
    nameN(i-1,n)

num = int(input())
nameN(num,num)
```

### backtracking

```python
def onetoN(i,n):
    if i > n:
        return 
    onetoN(i+1,n)
    print(i,end=" ")

num = int(input())
onetoN(1,num)
```
