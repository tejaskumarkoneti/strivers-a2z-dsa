```python

def countFreq(arr):

        n = len(arr)                     # Length of array

        # Create frequency array
        freq_count = [0] * (max(arr) + 1)

        # Count frequencies
        for num in arr:
            freq_count[num] += 1

        maxFreq = float("-inf")
        minFreq = float("inf")

        maxEle = -1
        minEle = -1

        # Traverse frequency array
        for i in range(len(freq_count)):

            # Ignore elements with 0 frequency
            if freq_count[i] > 0:

                # Update highest frequency
                if freq_count[i] > maxFreq:
                    maxFreq = freq_count[i]
                    maxEle = i

                # Update lowest frequency
                if freq_count[i] < minFreq:
                    minFreq = freq_count[i]
                    minEle = i

        print("The highest frequency element is:", maxEle)
        print("The lowest frequency element is:",minEle )
        
arr = [10, 5, 10, 15, 10, 5]
countFreq(arr)
```
