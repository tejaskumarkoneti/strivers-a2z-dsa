# Hashing Notes (DSA)

Hashing is one of the most important concepts in Data Structures & Algorithms.
It helps in fast storing and fetching of data.

## 1. What is Hashing?

**Hashing = Pre-storing + Fast Fetching**

## 2. Problem Understanding

Given an array:
```python
[1, 2, 1, 3, 2]
```
Queries:
- How many times does 1 appear?
- How many times does 2 appear?
- How many times does 4 appear?

## 3. Brute Force Approach

For every query:
- Traverse the entire array
- Count occurrences

````python
arr = [1, 2, 1, 3, 2]

def count_frequency(arr, number):
    count = 0

    for element in arr:
        if element == number:
            count += 1

    return count


print(count_frequency(arr, 1))
print(count_frequency(arr, 2))
print(count_frequency(arr, 4))
````

## Time Complexity

- One query → `O(N)`
- Q queries → `O(Q × N)`

## 4. Number Hashing

````python
arr = [1, 2, 1, 3, 2]

size = 13
hash_arr = [0] * size

# Pre storing
for num in arr:
    hash_arr[num] += 1

print(hash_arr)    

# Fast Fetching
queries = [1, 4, 2, 3, 12]

for q in queries:
    print(f"{q} appears {hash_arr[q]} times")
````

## 5. Character Hashing

````python
s = "abacbc"

hash_arr = [0] * 26

for ch in s:
    index = ord(ch) - ord('a')
    hash_arr[index] += 1

queries = ['a', 'b', 'c', 'z']

for ch in queries:
    index = ord(ch) - ord('a')
    print(f"{ch} appears {hash_arr[index]} times")
````

````python
#Code if the string contains both uppercase and lowercase:

# Input string
s = input("Enter the string")
print(s)

# Precompute frequency array
hash_arr = [0] * 256

for ch in s:
    hash_arr[ord(ch)] += 1

# Number of queries
q = int(input("Number of queries"))

# Fetch answers
for _ in range(q):
    ch = input()

    print("{} frequnecy in {}".format(ch,hash_arr[ord(ch)]))
````

## 6. Dictionary Hashing

````python
arr = [1, 2, 1, 3, 2]

freq = {}

for num in arr:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)

queries = [1, 2, 4]

for q in queries:
    print(f"{q} appears {freq.get(q, 0)} times")
````

## 7. Using Counter

````python
from collections import Counter

arr = [1, 2, 1, 3, 2]

freq = Counter(arr)

print(freq)

print(freq[1])
print(freq[2])
print(freq[4])
````

## 8. Collision in Hashing

Example:
```python
28 % 10 = 8
38 % 10 = 8
48 % 10 = 8
```
This is called a collision.

## 9. Highest Frequency Element

````python
arr = [1, 2, 1, 3, 2, 1]

freq = {}

for num in arr:
    freq[num] = freq.get(num, 0) + 1

max_freq = 0
answer = -1

for key, value in freq.items():
    if value > max_freq:
        max_freq = value
        answer = key

print(answer)
````

## Final Summary

- Hashing = prestoring + fetching
- Dictionary lookup is average `O(1)`
- Array hashing works for small ranges
- Character hashing uses ASCII
- `Counter` is built-in hashing utility
- Frequency problems → think hashing first
