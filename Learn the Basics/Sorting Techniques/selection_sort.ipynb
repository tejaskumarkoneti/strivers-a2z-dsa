# Selection Sort — Observation First Understanding

Think like this:

> “At every step, place the **smallest remaining element** in its correct position.”

That’s the whole idea of Selection Sort.


# Example

Array:

```python
[7, 5, 9, 2, 8]
```


# Core Observation

The array has two parts:

| Sorted Part | Unsorted Part |
|---|---|
| Left side | Right side |

Initially:

```python
Sorted:   []
Unsorted: [7, 5, 9, 2, 8]
```


# Iteration 1

Range:

```python
[7, 5, 9, 2, 8]
```

Minimum = `2`

Swap `2` with first element `7`

```python
[2, 5, 9, 7, 8]
```


# Iteration 2

Search minimum in:

```python
[5, 9, 7, 8]
```

Minimum = `5`

Already at correct position.

```python
[2, 5, 9, 7, 8]
```


# Iteration 3

Search minimum in:

```python
[9, 7, 8]
```

Minimum = `7`

Swap with `9`

```python
[2, 5, 7, 9, 8]
```


# Iteration 4

Search minimum in:

```python
[9, 8]
```

Minimum = `8`

Swap with `9`

```python
[2, 5, 7, 8, 9]
```

Array sorted ✅


# Real Logic Behind the Loops

## Outer Loop

```python
for i in range(n-1)
```

Purpose:

> Decide the current position where the correct smallest value should go.

---

## Inner Loop

```python
for j in range(i+1, n)
```

Purpose:

> Search for the minimum element in the unsorted part.


# Python Code

````
arr = [13, 46, 24, 52, 20, 9]

n = len(arr)

for i in range(n - 1):

    min_index = i

    for j in range(i + 1, n):

        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print(arr)
````

# Output

```python
[9, 13, 20, 24, 46, 52]
```


# Time Complexity

- Best Case: `O(n²)`
- Average Case: `O(n²)`
- Worst Case: `O(n²)`

Because we still search the remaining array every time.
