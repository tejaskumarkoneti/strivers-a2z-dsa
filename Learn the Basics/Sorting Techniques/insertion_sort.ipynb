# Insertion Sort — Simple Explanation

Insertion Sort works like arranging playing cards in your hand.

- Assume the **first element is already sorted**
- Pick the next element
- Place it in the correct position in the sorted part
- Repeat until the array is sorted


# Example

Input:

```python
[7, 4, 1, 5, 3]
```


## Step-by-Step Dry Run

### Step 1
Take `4`

Compare with `7`

`4 < 7` → shift `7`

Insert `4`

```python
[4, 7, 1, 5, 3]
```

---

### Step 2
Take `1`

Shift `7`
Shift `4`

Insert `1`

```python
[1, 4, 7, 5, 3]
```

---

### Step 3
Take `5`

Shift `7`

Insert `5`

```python
[1, 4, 5, 7, 3]
```

---

### Step 4
Take `3`

Shift `7`
Shift `5`
Shift `4`

Insert `3`

```python
[1, 3, 4, 5, 7]
```


# Simple Python Code

````

def insertion_sort(arr):

    # Traverse from second element
    for i in range(1, len(arr)):

        current = arr[i]
        j = i - 1

        # Shift bigger elements to the right
        while j >= 0 and arr[j] > current:
            arr[j + 1] = arr[j]
            j -= 1

        # Place current element at correct position
        arr[j + 1] = current

    return arr


# Driver Code
nums = [13, 46, 24, 52, 20, 9]

print("Before Sorting:")
print(nums)

sorted_array = insertion_sort(nums)

print("\nAfter Sorting:")
print(sorted_array)
````

# Important Notes

## Observation

Insertion Sort divides array into:

- **Sorted Part**
- **Unsorted Part**

At every iteration:
- Pick one element from unsorted part
- Insert it into correct position in sorted part


# Time Complexity

| Case | Complexity |
|---|---|
| Best Case | O(N) |
| Average Case | O(N²) |
| Worst Case | O(N²) |


# Space Complexity

```python
O(1)
```

Because sorting happens inside same array.


# Best Case

When array is already sorted:

```python
[1, 2, 3, 4, 5]
```

Only comparisons happen.

No shifting.


# Worst Case

When array is reverse sorted:

```python
[5, 4, 3, 2, 1]
```

Every element shifts multiple times.


# Key Idea

For every element:

```text
Can this element go left?
YES → Shift bigger elements
NO  → Stop
```


# Interview Notes

## Why called “Insertion” Sort?

Because we **insert** each element into its correct position.

---

## Stable Sorting?

✅ Yes

---

## In-place Sorting?

✅ Yes

No extra array used.
