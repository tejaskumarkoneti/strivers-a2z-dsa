# Bubble Sort Algorithm

Bubble Sort repeatedly compares adjacent elements and swaps them if they are in the wrong order

The largest element “bubbles” to the end after every pass.


# Brute Force Approach

## Idea

- Compare adjacent elements
- Swap if left element is greater
- Perform all passes even if array becomes sorted early


```
def bubble_sort(arr):

    n = len(arr)

    # Outer loop for passes
    for i in range(n - 1):

        # Inner loop for comparison
        for j in range(n - i - 1):

            # Swap if elements are in wrong order
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


arr = [13, 46, 24, 52, 20, 9]

print("Before Using Bubble Sort:")
print(arr)

bubble_sort(arr)

print("After Using Bubble Sort:")
print(arr)
```

# Complexity Analysis (Brute Force)

| Case | Time Complexity |
|---|---|
| Best Case | O(N²) |
| Average Case | O(N²) |
| Worst Case | O(N²) |

### Space Complexity
O(1)


# Optimal Approach

## Main Observation

If during one full pass:

- no swapping occurs
- array is already sorted

So we can stop early using a boolean flag.


```
def optimized_bubble_sort(arr):

    n = len(arr)

    # Outer loop
    for i in range(n - 1, -1, -1):

        did_swap = False

        # Inner loop
        for j in range(i):

            # Swap if needed
            if arr[j] > arr[j + 1]:

                arr[j], arr[j + 1] = arr[j + 1], arr[j]

                did_swap = True

        # If no swap happens, array already sorted
        if not did_swap:
            break


arr = [13, 46, 24, 52, 20, 9]

print("Before Using Optimized Bubble Sort:")
print(arr)

optimized_bubble_sort(arr)

print("After Using Optimized Bubble Sort:")
print(arr)
```

# Complexity Analysis (Optimal)

| Case | Time Complexity |
|---|---|
| Best Case | O(N) |
| Average Case | O(N²) |
| Worst Case | O(N²) |

### Space Complexity
O(1)


# Key Interview Understanding

Bubble Sort works because:

After every pass:
- the largest unsorted element reaches its correct position

Example:

- Pass 1 → largest element fixed
- Pass 2 → second largest fixed
- Pass 3 → third largest fixed

Sorted portion grows from the right side.
