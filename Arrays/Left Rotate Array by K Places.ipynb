### Brute force

```python
def rotateD(arr,d):


    n = len(arr)

    d = d % n
    
    res = [0]*n
    
    for i in range(n - d):

        res[i] = arr[d+i]

    for i in range(d):

        res[n-d+i] = arr[i]

    return res


r = rotateD([1,2,3,4,5,6,7,9],17)
print(r)
```

```python
arr = [1,2,3,4,5,6,7]
k = 3
n = len(arr)
temp = []

for  i in range(k):

    temp.append(arr[i])

print(temp)

for i in range(k,n):
    arr[i-k] = arr[i]

print(arr)

for i in range(n-k,n):

    arr[i] = temp[i - (n-k)]

print(arr)
```

### optimal

```python
def reverse(arr,p1,p2):

    while p1 < p2:

        arr[p1],arr[p2] = arr[p2],arr[p1]

        p1 += 1
        p2 -= 1
    
    return arr
arr = [1,2,3,4,5,6]
n = len(arr)
k = 2
reverse(arr,0,k-1)
reverse(arr,k,n-1)
reverse(arr,0,n-1)


```
