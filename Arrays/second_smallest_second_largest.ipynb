# Find Second Smallest and Second Largest Element in an Array

## Problem Statement

Given an array, find:

- Second Smallest Element
- Second Largest Element

If either does not exist, return `-1`.

---

## Example

### Input

```python
arr = [1, 2, 4, 7, 7, 5]
```

### Output

```python
Second Smallest = 2
Second Largest = 5
```


# Observation First 🧠

We need to track:

- smallest
- second_smallest
- largest
- second_largest

Instead of sorting, we observe every element once and update values carefully.


# Core Logic

## For Second Smallest

### Case 1

If current element is smaller than `small`:

```python
second_small = small
small = arr[i]
```

The old smallest becomes the second smallest.

---

### Case 2

If current element is:

- greater than `small`
- smaller than `second_small`

Then:

```python
second_small = arr[i]
```

# Why Duplicate Check is Important ⚠️

Example:

```python
arr = [7, 7, 7]
```

There is no second largest distinct element.

That is why we use:

```python
arr[i] != large
```

and

```python
arr[i] != small
```


# Optimal Code (O(N))

Single traversal style.


````
def secondSmallest(arr, n):

    if n < 2:
        return -1

    small = float('inf')
    second_small = float('inf')

    for i in range(n):

        if arr[i] < small:

            second_small = small
            small = arr[i]

        elif arr[i] < second_small and arr[i] != small:

            second_small = arr[i]

    if second_small == float('inf'):
        return -1

    return second_small


def secondLargest(arr, n):

    if n < 2:
        return -1

    large = float('-inf')
    second_large = float('-inf')

    for i in range(n):

        if arr[i] > large:

            second_large = large
            large = arr[i]

        elif arr[i] > second_large and arr[i] != large:

            second_large = arr[i]

    if second_large == float('-inf'):
        return -1

    return second_large


arr = [1, 2, 4, 7, 7, 5]

n = len(arr)

print("Second Smallest:", secondSmallest(arr, n))
print("Second Largest:", secondLargest(arr, n))
````

# Better Approach (Two Traversals)

## Algorithm

### Step 1

Traverse the array once to find:

- Smallest element
- Largest element

---

### Step 2

Traverse the array again:

- Find the element just greater than the smallest element  
  → Second Smallest

- Find the element just smaller than the largest element  
  → Second Largest

---

## Key Observation 🧠

We first identify the boundary values:

```python
smallest
largest
```

Then we search for the best candidates around them.


````
# Function to find the second smallest and second largest elements in the array
def getElements(arr, n):

    # Edge case
    if n == 0 or n == 1:
        print(-1, -1)
        return

    # Initialize variables
    small = float('inf')
    second_small = float('inf')

    large = float('-inf')
    second_large = float('-inf')

    # First traversal
    # Find smallest and largest
    for i in range(n):

        small = min(small, arr[i])
        large = max(large, arr[i])

    # Second traversal
    # Find second smallest and second largest
    for i in range(n):

        if arr[i] < second_small and arr[i] != small:
            second_small = arr[i]

        if arr[i] > second_large and arr[i] != large:
            second_large = arr[i]

    print("Second smallest is", second_small)
    print("Second largest is", second_large)


# Driver Code
arr = [1, 2, 4, 6, 7, 5]

n = len(arr)

getElements(arr, n)
````

# Complexity Analysis — Better Approach

## Time Complexity

```python
O(N)
```

Why?

- First traversal → O(N)
- Second traversal → O(N)

Total:

```python
O(N + N) = O(N)
```

---

## Space Complexity

```python
O(1)
```

Because only constant variables are used.
