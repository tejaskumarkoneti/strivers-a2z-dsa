
# Check if an Array is Sorted

## Problem
Given an array of size `n`, check whether the array is sorted in:

```text
Ascending / Non-decreasing order
```

Return:

```text
True  -> if sorted
False -> otherwise
```

---

# Example Input / Output

## Example 1

### Input
```python
arr = [1,2,3,4,5]
```

### Output
```python
True
```

### Explanation
Every element is:

```text
smaller than or equal to next element
```

So array is sorted.

---

## Example 2

### Input
```python
arr = [5,4,6,7,8]
```

### Output
```python
False
```

### Explanation
```python
5 > 4
```

which violates sorted order.

---

# Pattern

```text
Array Traversal + Adjacent Comparison
```

---

# Brute Force Approach

## Core Idea

For every element:
```text
compare with all future elements
```

If any future element is smaller:
```text
array is not sorted
```

---

# Brute Force Logic

```python
def isSorted(arr, n):

    for i in range(n):

        for j in range(i + 1, n):

            if arr[j] < arr[i]:
                return False

    return True
```

---

# Brute Force Complexity

| Complexity | Value |
|---|---|
| Time | O(N²) |
| Space | O(1) |

---

# Observation / Reasoning

We do NOT need to compare:

```text
every pair
```

because in a sorted array:

```text
current element >= previous element
```

If this condition holds for all adjacent pairs,
then entire array is sorted.

---

# Optimal Approach

## Key Observation

Check only:

```python
arr[i] >= arr[i-1]
```

for every index.

---

# Invariant

```text
Everything before current index remains sorted
```

---

# Core Logic

```python
def isSorted(arr, n):

    for i in range(1, n):

        if arr[i] < arr[i - 1]:
            return False

    return True
```

---

# Pointer / Condition Explanation

| Variable | Meaning |
|---|---|
| `i-1` | Previous element |
| `i` | Current element |

---

## Condition

```python
if arr[i] < arr[i-1]
```

Meaning:
```text
current element became smaller than previous
```

which breaks ascending order.

So:

```python
return False
```
