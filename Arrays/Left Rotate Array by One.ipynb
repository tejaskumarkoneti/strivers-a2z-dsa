### brute force approach

```python
nums = [1,2,3,4,5]

res = []

for i in range(1,len(nums)):

    res.append(nums[i])

res.append(nums[0])

print(res)
```

```python
nums = [1,2,3,4,5]

res = [0] * len(nums)
print(res)

temp = nums[0]

for i in range(1,len(nums)):

    res[i-1] = nums[i]

res[len(nums)-1] = temp

print(res)
```

### better


```python
nums = [1,2,3,4,5]

temp = nums[0]

for i in range(1,len(nums)):

    nums[i-1] = nums[i]

nums[len(nums)-1] = temp

print(nums)
```
