# Remove Duplicates from Sorted Array

## Approach 1 — Brute Force Using Set

### 1. Problem

Given a sorted array, remove duplicates such that every unique element appears only once.

---

### 2. Example Input / Output

#### Input
```python
arr = [1,1,2,2,2,3,3]
```

#### Output
```python
[1,2,3,_,_,_,_]
k = 3
```


## 3. Core Logic

Use a set because:

```python
set = unique values only
```

Steps:

1. Traverse array
2. Store unique values in set
3. Copy unique values back into beginning of array
4. Return count


## 4. Pointer / Condition Explanation

| Variable | Purpose |
|---|---|
| seen | stores unique elements |
| index | where next unique element should be placed |

Important condition:

```python
if num not in seen
```

Meaning:
- if already exists → duplicate
- else → unique


## 5. Observation / Reasoning

A set automatically removes duplicates.

Example:

```python
[1,1,2,2,3]
```

becomes:

```python
{1,2,3}
```

We overwrite the array because the problem asks us to modify the array in-place.


## 6. Tracing

Input:

```python
arr = [1,1,2,2,2,3,3]
```

### Step 1
- num = 1
- seen = {1}
- arr[0] = 1

### Step 2
- duplicate 1
- skip

### Step 3
- num = 2
- seen = {1,2}
- arr[1] = 2

Array becomes:

```python
[1,2,2,2,2,3,3]
```

### Step 4
- duplicate 2s skipped

### Step 5
- num = 3
- arr[2] = 3

Final:

```python
[1,2,3,_,_,_,_]
```


````
# Brute Force Code

class Solution:

    def removeDuplicates(self, nums):

        seen = set()

        index = 0

        for num in nums:

            if num not in seen:

                seen.add(num)

                nums[index] = num

                index += 1

        return index


nums = [1,1,2,2,2,3,3]

sol = Solution()

k = sol.removeDuplicates(nums)

print(nums[:k])
print(k)
````

## Complexity

| Complexity | Value |
|---|---|
| Time | O(N) |
| Space | O(N) |


# Approach 2 — Optimal Two Pointer Approach

### 1. Problem

Remove duplicates from sorted array in-place using minimum extra space.


## 2. Example Input / Output

#### Input
```python
arr = [1,1,2,2,2,3,3]
```

#### Output
```python
[1,2,3,_,_,_,_]
k = 3
```


## 3. Core Logic

Since array is sorted:

```python
duplicates stay together
```

Use:
- i → last unique element
- j → scans array


## 4. Pointer / Condition Explanation

| Pointer | Purpose |
|---|---|
| i | points to last unique element |
| j | scans every element |

Condition:

```python
if nums[j] != nums[i]
```

Meaning:
- new unique element found


## 5. Observation / Reasoning

Because the array is sorted:

```python
1 1 1 2 2 3
```

Duplicates stay adjacent.

So whenever value changes:

```python
1 -> 2
```

we found a new unique number.


## 6. Tracing

Input:

```python
arr = [1,1,2,2,2,3,3]
```

Initial:

```python
i = 0
j = 1
```

Array:

```python
[1,1,2,2,2,3,3]
 i j
```

### Step 1
1 == 1 → duplicate → skip

### Step 2
j = 2

1 != 2 → unique found

Move i:

```python
i = 1
```

Place:

```python
nums[1] = 2
```

Array:

```python
[1,2,2,2,2,3,3]
```

### Step 3
Skip remaining duplicate 2s

### Step 4
3 found → place at next unique position

Final:

```python
[1,2,3,_,_,_,_]
```


````
# Optimal Code

class Solution:

    def removeDuplicates(self, nums):

        if not nums:
            return 0

        i = 0

        for j in range(1, len(nums)):

            if nums[j] != nums[i]:

                i += 1

                nums[i] = nums[j]

        return i + 1


nums = [1,1,2,2,2,3,3]

sol = Solution()

k = sol.removeDuplicates(nums)

print(nums[:k])
print(k)
````

## Complexity

| Complexity | Value |
|---|---|
| Time | O(N) |
| Space | O(1) |
