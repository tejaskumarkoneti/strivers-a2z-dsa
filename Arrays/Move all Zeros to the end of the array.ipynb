```python
arr = [1,0,2,3,0,4,0,1]

res = []

n = len(arr)

for i in range(n):

    if arr[i] != 0:

        res += [arr[i]]

print(res)

for i in range(len(res), len(arr)):

    res += [0]

print(res)
```

```python
arr = [1,0,2,3,0,4,0,1]

result = []

n = len(arr)
print(n)
# non - zero element in result array
for i in range(n):

    if arr[i] != 0:

        result += [arr[i]]

print(result)
nz = len(result)
print(nz)

# shifting non-zero ele back to orginal array
for i in range(nz):

    arr[i] = result[i]

print(arr)

# adding zeros after non-zero elements in orginal array
for i in range(nz, n):

    arr[i] = 0

print(arr)
```

### Optimal 


```python
arr = [1,0,2,3,0,4,0,1]

n = len(arr)

for i in range(n):

    if arr[i] == 0:

        j = i
        break


for i in range(j+1,n):

    if arr[i] != 0:

        arr[i],arr[j] = arr[j],arr[i]
        j+=1

print(arr)
```
